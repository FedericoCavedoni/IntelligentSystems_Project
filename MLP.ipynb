```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from pandas import read_csv
from numpy import ravel
from sklearn.model_selection import GridSearchCV
import random
import numpy as np
import random
```

```python
data = read_csv('/content/drive/Shareddrives/Di Ricco_Barbieri_Cavedoni/whole_data_15.csv')
data = data.sample(frac=1, random_state=random.seed())

training_data, testing_data, training_labels, testing_labels = train_test_split(data.iloc[:, 7:len(data.columns)], data.iloc[:, 1])
```

**Esecuzione senza Early Stopping**

```python
mlp = MLPClassifier(verbose=True)
mlpmodel = mlp.fit(training_data, ravel(training_labels))
mlplabels = mlpmodel.predict(testing_data)
mlpscore = accuracy_score(ravel(testing_labels), mlplabels)
print("MPLClassifier score:", mlpscore)
```

**Esecuzione con Early Stopping**

```python
mlp = MLPClassifier(validation_fraction=0.2, early_stopping=True, verbose=True)
mlpmodel = mlp.fit(training_data, ravel(training_labels))
mlplabels = mlpmodel.predict(testing_data)
mlpscore = accuracy_score(ravel(testing_labels), mlplabels)
print("MPLClassifier score:", mlpscore)
```

**TUNING MLP**

```python
# Carica il dataset di esempio
data = read_csv('/content/drive/Shareddrives/Di Ricco_Barbieri_Cavedoni/whole_data_15.csv')
data = data.sample(frac=1, random_state=random.seed())


training_data, testing_data, training_labels, testing_labels = train_test_split(data.iloc[:, 7:len(data.columns)], data.iloc[:, 1])

# Definisci la griglia di valori per gli iperparametri
param_grid = {
    'hidden_layer_sizes': [200, 500, 1000],
    'activation': ['relu', 'tanh'],
    'solver': ['sgd', 'adam'],
    'learning_rate_init': [0.001, 0.01, 0.1],
    'max_iter': [100,200,500],
}

# Crea l'oggetto MLPClassifier
mlp = MLPClassifier(validation_fraction=0.2, early_stopping=True)

# Esegui la grid search
grid_search = GridSearchCV(mlp, param_grid, cv=3)
grid_search.fit(training_data, training_labels)

# Stampa i risultati della grid search
print("Migliori iperparametri:", grid_search.best_params_)
print("Migliore accuratezza:", grid_search.best_score_)
```
